Algoritmo que Recebe um PDF como entrada, analisa a frequência das palavras e gera uma WordCloud com as palavras mais relevantes.

```python
!pip install pypdf2
!pip install wordcloud

import PyPDF2
from PyPDF2 import PdfReader

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
pdffileobj=open('pdf_unidade2.pdf','rb')
pdfreader=PyPDF2.PdfReader(pdffileobj)

x = len(pdfreader.pages)

pageobj= pdfreader.pages[0]
text = pageobj.extract_text()
```

```python
# creating a pdf reader object
reader = PdfReader('pdf_unidade2.pdf')

# printing number of pages in pdf file
x = (len(reader.pages))
```

```python
texto = ""
for i in range(3,x):
  page = reader.pages[i]
  text = page.extract_text()
  if ("ENCERRA AQUI O TRECHO DO LIVRO" or "O conteúdo deste livro") in text:
    continue
  else:
    texto = " ".join([texto, text])

texto = texto.lower()
```

```python
stopwords = set(STOPWORDS)
stopwords.update(['objetivos','com','do','da','em','no','na','o','a','é','e','os','as','um','uma','nos','nas',
             'há', 'p.','ser','parte','unidade', 'conteúdo','introdução','qual','quais','conforme','então','assim',
             'de','que','dos','para','como','figura','devem','seu','ter','sendo','pelo','mais','pela','por','além disso',
             'irá','forma','deve','sobre','pois','são','além','seus','suas','à','está','autores','outro','fonte','podem',
             'sempre','alguns','quanto','todo','se','sua','dessa','ponto','outra','das','entre','br','ele','ela','quando',
             'cada','até','título','titulo','tem','principais','também','ago','essa','você','podemos','fazer','p','ainda','melhor',
             'nova','meio','uso','pode','toda','sejas','tipo','fernandes','base','desse','já','melhor','abreu','ou','às','utilizado',
             'fernandes abreu','disponível', 'acordo','não','esse','outros','ao','aos','seja','todos','contexto','dentro','informações',
             'melhorar','estratégico','brasil','regulamentaçõe', 'acesso'])

# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="white",
                      width=1600, height=800, max_words=30,
                      collocations=False).generate(texto)
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.imshow(wordcloud);
```
